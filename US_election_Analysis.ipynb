# there is electrol college system to assign seat to winning party so, import statewise number of electrol college csv file (created By me using web page import)

```python
df_ec = pd.read_csv('state_ec.csv')
```

```python
df_ec
```

```python
df_2020 = df[df['year']==2020].groupby(['state','state_po','party_simplified'], as_index=False)['candidatevotes'].sum()
df_2020
```

```python
df_2020_max = df_2020.groupby(['state','state_po'], as_index=False).apply(lambda x: x.nlargest(1,['candidatevotes'])).reset_index(drop=True)
df_2020_max
```

```python
new_df = pd.merge(df_2020_max, df_ec, on='state_po')
new_df
```

```python
win = new_df.groupby('party_simplified',as_index=False)['ec'].sum()
win
```

```python
win.loc[win.ec == max(win.ec),'party_simplified']
```

```python
def us_results(yr):
    df_yr = df[df['year']==yr].groupby(['state','state_po','party_simplified'], as_index=False)['candidatevotes'].sum()
    df_yr_max = df_yr.groupby(['state','state_po'], as_index=False).apply(lambda x: x.nlargest(1,['candidatevotes'])).reset_index(drop=True)
    new_df = pd.merge(df_yr_max, df_ec, on='state_po')
    win = new_df.groupby('party_simplified',as_index=False)['ec'].sum()
    print(win.loc[win.ec == max(win.ec),'party_simplified'])
```

```python
us_results(2020)
## correct confirmed with Goggle 
```
